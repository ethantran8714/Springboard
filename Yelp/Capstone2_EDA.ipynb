### Correlation Matrix Among Dataset

## Since we couldn't find a correlation with the explicit variables, we look to the words to see if there are any patterns we could point out.

### To extend our analysis, we wanted to look at the sentiment of text reviews. 

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


analyzer = SentimentIntensityAnalyzer()
df_neg = pd.Series(dtype="float64")
df_pos = pd.Series(dtype="float64")
df_neu = pd.Series(dtype="float64")
df_compound = pd.Series(dtype="float64")

for row in df.itertuples():
    vs = analyzer.polarity_scores(row.Text)
    df_neg = df_neg.append(pd.Series(vs["neg"]), ignore_index=True)
    df_pos = df_pos.append(pd.Series(data = vs["pos"]), ignore_index=True)
    df_neu = df_neu.append(pd.Series(data = vs["neu"]), ignore_index=True)
    df_compound = df_compound.append(pd.Series(data = vs["compound"]), ignore_index=True)

df["negative_sentiment"] = df_neg
df["positve_sentiment"] = df_pos
df["neutral_sentiment"] = df_neu
df["composite_sentiment"] = df_compound
```

```python
df.head()
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
hm = sns.heatmap(round(corrmat,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Feature Correlation for All Reviews', fontsize=14)
```

#### With our newfound sentiment scores, we see a few trends:

Stars have a moderately positive correlation with composite and positive sentiment.
Stars have a moderately negative correlation with negative and neutral sentiment.

Positive sentiment has a positive correlation with composite sentiment.
Positive sentiment has a negative correlation with negative and neutral sentiment.
